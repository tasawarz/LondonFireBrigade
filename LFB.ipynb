```python
import pyproj

# Define the BNG CRS
bng = pyproj.Proj(proj='tmerc', ellps='airy', datum='OSGB36',
                  lat_0=49, lon_0=-2, x_0=400000, y_0=-100000)

# Define the WGS 84 CRS (used for latitude and longitude)
wgs84 = pyproj.Proj(proj='latlong', datum='WGS84')

# Define your Easting and Northing coordinates
easting = data['Easting_rounded']
northing = data['Northing_rounded']

# Convert the coordinates
longitude, latitude = pyproj.transform(bng, wgs84, easting, northing)

print("Longitude:", longitude)
print("Latitude:", latitude)

data['Longitude'] = longitude
data['Latitude'] = latitude
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' contains the dataset with the required columns and data loaded

# Convert 'DateOfCall' and 'TimeOfCall' columns into datetime format
data['DateTimeOfCall'] = pd.to_datetime(data['DateOfCall'] + ' ' + data['TimeOfCall'])

# Extract temporal information
data['Year'] = data['DateTimeOfCall'].dt.year
data['Month'] = data['DateTimeOfCall'].dt.month
data['DayOfWeek'] = data['DateTimeOfCall'].dt.dayofweek
data['HourOfDay'] = data['DateTimeOfCall'].dt.hour

# Temporal analysis - Incidents per year
incidents_per_year = data['Year'].value_counts().sort_index()

# Temporal analysis - Incidents per month in a specific year (e.g., 2018)
incidents_per_month_2018 = data[data['Year'] == 2018]['Month'].value_counts().sort_index()

# Temporal analysis - Incidents per day of the week
day_of_week_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
incidents_per_day_of_week = data['DayOfWeek'].value_counts().sort_index()

# Temporal analysis - Incidents per hour of the day
incidents_per_hour_of_day = data['HourOfDay'].value_counts().sort_index()

# Create graphs

# Incidents per year
plt.figure(figsize=(10, 6))
plt.bar(incidents_per_year.index, incidents_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.title('Incidents per Year')
plt.show()

# Incidents per month in 2018
plt.figure(figsize=(10, 6))
plt.plot(incidents_per_month_2018.index, incidents_per_month_2018.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.title('Incidents per Month in 2018')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

# Incidents per day of the week
plt.figure(figsize=(10, 6))
plt.bar(day_of_week_names, incidents_per_day_of_week.values)
plt.xlabel('Day of the Week')
plt.ylabel('Number of Incidents')
plt.title('Incidents per Day of the Week')
plt.show()

# Incidents per hour of the day
plt.figure(figsize=(10, 6))
plt.plot(incidents_per_hour_of_day.index, incidents_per_hour_of_day.values, marker='o')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Incidents')
plt.title('Incidents per Hour of the Day')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Assuming 'df' contains the dataset with the required columns and data loaded

# Convert Longitude and Latitude to a GeoDataFrame for spatial analysis
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['Longitude'], data['Latitude']))

# Spatial Analysis - Plot incidents on a map
fig, ax = plt.subplots(figsize=(12, 8))
gdf.plot(ax=ax, markersize=10, color='red', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Fire Incidents')
plt.show()

# Spatial Analysis - Incidents per Borough
borough_counts = data['IncGeo_BoroughName'].value_counts()

plt.figure(figsize=(12, 6))
borough_counts.plot(kind='bar')
plt.xlabel('Borough')
plt.ylabel('Number of Incidents')
plt.title('Incidents per Borough')
plt.xticks(rotation=45, ha='right')
plt.show()

# Spatial Analysis - Incidents per Ward
ward_counts = data['IncGeo_WardNameNew'].value_counts()

plt.figure(figsize=(14, 6))
ward_counts.plot(kind='bar')
plt.xlabel('Ward')
plt.ylabel('Number of Incidents')
plt.title('Incidents per Ward')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Assuming 'df' contains the dataset with the required columns and data loaded

# Time Series Analysis - Incidents over Time
# Resample the data by month and count incidents per month
incidents_per_month = data.resample('M')['IncidentNumber'].count()

# Plotting the time series of incidents over time
plt.figure(figsize=(12, 6))
plt.plot(incidents_per_month.index, incidents_per_month.values, marker='o')
plt.xlabel('Date')
plt.ylabel('Number of Incidents')
plt.title('Time Series Analysis: Incidents over Time')
plt.grid(True)
plt.show()

# Correlation Analysis
# Let's consider numerical columns for correlation analysis
numerical_columns = ['FirstPumpArriving_AttendanceTime', 'NumPumpsAttending', 'PumpHoursRoundUp']

# Create a correlation matrix for numerical columns
correlation_matrix = data[numerical_columns].corr()

# Visualize correlation using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Analysis: Numerical Columns')
plt.show()

# Regression Analysis
# We'll use ordinary least squares (OLS) regression to analyze 'NumPumpsAttending' based on 'FirstPumpArriving_AttendanceTime' and 'PumpHoursRoundUp'
X = data[['FirstPumpArriving_AttendanceTime', 'PumpHoursRoundUp']]
X = sm.add_constant(X)  # Add a constant term for the intercept
y = data['NumPumpsAttending']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression model
print(model.summary())
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.impute import SimpleImputer

# Assuming 'df' is the DataFrame containing the data with the required columns

# Define predictor variables and the dependent variable
predictors = ['NumStationsWithPumpsAttending', 'PumpCount', 'FirstPumpArriving_AttendanceTime', 'SecondPumpArriving_AttendanceTime']
X = data[predictors]
y = data['NumPumpsAttending']

# Use SimpleImputer for imputation
imputer = SimpleImputer(strategy='mean')

# Impute missing values
X_imputed = imputer.fit_transform(X)

# Convert the imputed array back to DataFrame
X_imputed_df = pd.DataFrame(X_imputed, columns=predictors)

# Add a constant term to the predictor variables for the intercept term in the regression
X_imputed_df = sm.add_constant(X_imputed_df)

# Reset the index for the y Series
y = y.reset_index(drop=True)

# Fit the regression model
model = sm.OLS(y, X_imputed_df).fit()

# Analyze the regression results
print(model.summary())
```
